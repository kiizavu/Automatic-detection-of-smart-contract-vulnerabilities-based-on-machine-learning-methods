```python
import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Lambda
from keras.layers import Conv2D, MaxPooling2D
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.optimizers import Adam
from sklearn.utils import shuffle
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
import time
from keras.utils import to_categorical
from imblearn.over_sampling import SMOTE 
```

```python
df_1gram = pd.read_csv("contract_labled_1_1-gram.csv")
df_2gram = pd.read_csv("contract_labled_1_2-gram.csv")
df_3gram = pd.read_csv("contract_labled_1_3-gram.csv")
df_4gram = pd.read_csv("contract_labled_1_4-gram.csv")
df_5gram = pd.read_csv("contract_labled_1_5-gram.csv")
df_6gram = pd.read_csv("contract_labled_1_6-gram.csv")
df_7gram = pd.read_csv("contract_labled_1_7-gram.csv")
df_8gram = pd.read_csv("contract_labled_1_8-gram.csv")
df_9gram = pd.read_csv("contract_labled_1_9-gram.csv")
df_10gram = pd.read_csv("contract_labled_1_10-gram.csv")
```

```python
def DNN(df):
    df = df.fillna(0)
    shuffle(df)
    data = df.values

    #Trích xuất data và label
    X = df.iloc[:,6:].values
    y = df[['Label']].values
    y = to_categorical(y)
    #nhãn 1, 0
    X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 21, test_size=0.2)
    #Chuẩn hóa data
    #z = (x – u ) / s
    # Trong đó:
    # z : là giá trị mới.
    # x : là giá trị gốc ban đầu.
    # u : là giá trị trung bình (mean).
    # s : là độ lệch chuẩn.
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    inputs = X_train[0].shape
    num_classes = y_train.shape[1]

    model = Sequential()
    model.add(Dense(512, input_shape=(inputs), activation='relu')) #Inputs
    model.add(Dense(256, activation='relu')) #công thức f(x)=max(0,x)
    model.add(Dropout(0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='sigmoid'))
    #Sigmoid nó sẽ trả về (0,1)
    model.compile(loss='binary_crossentropy',
                optimizer='adam',
                metrics=['binary_accuracy'])

    model.summary()
    import time

    start_train = time.time()

    test = model.fit(x = X_train, y = y_train, epochs=10, validation_data=(X_test, y_test))

    end_train = time.time()

    time_train = end_train - start_train
    start_pred = time.time()

    #thực hiện Predict
    y_pred = model.predict(X_test)

    y_pred_class=model.predict(X_test) 
    y_pred_class =np.argmax(y_pred_class ,axis=1)

    from sklearn.metrics import confusion_matrix
    y_pred = model.predict(X_test)
    y_test_class = np.argmax(y_test)

    end_pred = time.time()
    time_pred = end_pred - start_pred
    print(y_pred_class)
    
    #Print loss and acc
    print("Time Train: ",time_train)
    print("Time predict: ", time_pred)
    loss, accuracy = model.evaluate(X_test, y_test, verbose = 0)
    print('Model Loss: %.2f, Accuracy: %.2f' %((loss), (accuracy)))
    matrix = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))

    print(matrix)

    print("\n classification_report: ")
    print(classification_report(y_test.argmax(axis=1), y_pred.argmax(axis=1)))
    
    
```

```python
def DNN_SM(df):
    df = df.fillna(0)
    shuffle(df)
    data = df.values

    #Trích xuất data và label
    X = df.iloc[:,6:].values
    y = df[['Label']].values
    y = to_categorical(y)
    sm = SMOTE(random_state = 2)
    X, y = sm.fit_resample(X, y)
    #nhãn 1, 0
    X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 21, test_size=0.2)

    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    inputs = X_train[0].shape
    num_classes = y_train.shape[1]

    model = Sequential()
    model.add(Dense(512, input_shape=(inputs), activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='sigmoid'))

    model.compile(loss='binary_crossentropy',
                optimizer='adam',
                metrics=['binary_accuracy'])

    model.summary()
    import time

    start_train = time.time()

    test = model.fit(x = X_train, y = y_train, epochs=10, validation_data=(X_test, y_test))

    end_train = time.time()

    time_train = end_train - start_train
    start_pred = time.time()

    #thực hiện Predict
    y_pred = model.predict(X_test)

    y_pred_class=model.predict(X_test) 
    y_pred_class =np.argmax(y_pred_class ,axis=1)

    from sklearn.metrics import confusion_matrix
    y_pred = model.predict(X_test)
    y_test_class = np.argmax(y_test)

    end_pred = time.time()
    time_pred = end_pred - start_pred
    print(y_pred_class)
    
    #Print loss and acc
    print("Time Train: ",time_train)
    print("Time predict: ", time_pred)
    loss, accuracy = model.evaluate(X_test, y_test, verbose = 0)
    print('Model Loss: %.2f, Accuracy: %.2f' %((loss), (accuracy)))
    matrix = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))

    print(matrix)

    print("\n classification_report: ")
    print(classification_report(y_test.argmax(axis=1), y_pred.argmax(axis=1)))
```
